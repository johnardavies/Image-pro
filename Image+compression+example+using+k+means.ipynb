## Image compression example using k means

This notebook applies k means to a picture's Blue Green Red (BGR) channels to approximate the picture using k pixel BGR combinations

```python
import numpy as np
import cv2
from numpy import *
import os
import csv


#Applies it to a single picture
pathlist="filepath"

#reads in the photograph
read_original=cv2.imread(pathlist)
img=array(read_original) 

#Plots the original picture
imshow(img)
arr = np.float32(img)

#reshapes the photo data from X, Y, 3 to X*Y, 3 i.e 3 columns and X*Y rows. Opwn cv stores the colors in the BGR format
pixels = arr.reshape((-1, 3))
#The number of clusters
n_colors = 3
 # sets up the criteria to end the k means algorithm
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 200, .1)
#This defines how the initial centres are initated
flags = cv2.KMEANS_RANDOM_CENTERS

#Runs the k means analysis
_, labels, centroids = cv2.kmeans(pixels, n_colors, criteria, 10, flags)

```

```python
store=[]
for elem in labels:
  for i in range(0,n_colors):
   if elem==[i]:
      store.append(centroids[i].tolist())
```

```python
import numpy as np
import matplotlib
from pylab import *
%matplotlib inline


#Converts the nested lists to an array
a=array(store)
#Reshapes the data to its original dimenions
b=a.reshape(640, 640,3)

#Plots the image
imshow(b)
```
